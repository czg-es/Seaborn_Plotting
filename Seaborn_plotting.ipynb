# Bar plot 
Muestra estimaciones puntuales e intervalos de confianza como barras rectangulares.
Tambin es importante tener en cuenta que un grfico de barras muestra solo el valor medio (u otro estimador),
pero en muchos casos puede ser ms informativo mostrar la distribucin de valores en cada nivel de las variables categricas.
En ese caso, otros enfoques, como una trama de caja o violn, pueden ser ms apropiados.

<!--
Show point estimates and confidence intervals as rectangular bars.
It is also important to keep in mind that a bar plot shows only the mean (or other estimator) value,
but in many cases it may be more informative to show the distribution of values at each level of the categorical variables.
In that case, other approaches such as a box or violin plot may be more appropriate.
-->

````python
plt.figure(figsize=(15,5),facecolor='#00000000')
sns.barplot(x = 'Age', y = 'Sex', data = datanew, estimator= np.mean); 
````

# Count plot

Muestra el conteo de observaciones en cada contenedor categrico usando barras.
Un grfico de recuento se puede considerar como un histograma a travs de una variable categrica, en lugar de cuantitativa.
Esta funcin siempre trata una de las variables como categrica y extrae datos
en posiciones ordinales (0, 1, ... n) en el eje correspondiente,
incluso cuando los datos tienen un tipo numrico o de fecha.
<!--
Show the counts of observations in each categorical bin using bars.
A count plot can be thought of as a histogram across a categorical, instead of quantitative, variable. 
This function always treats one of the variables as categorical and draws data 
at ordinal positions (0, 1,  n) on the relevant axis,
even when the data has a numeric or date type.
-->

````python
plt.figure(figsize=(15,5),facecolor='#00000000')
sns.countplot(x = 'Pclass', data = datanew);
````

# Box plot
Un diagrama de caja (o diagrama de caja y bigotes) muestra la distribucin de datos cuantitativos
de una manera que facilite las comparaciones entre variables o entre niveles de una variable categrica.
El cuadro muestra los cuartiles del conjunto de datos mientras que los bigotes se extienden para mostrar el resto de la distribucin,
excepto para los puntos que se determinan como "valores atpicos" utilizando un mtodo que es una funcin del rango intercuartil.
<!--
A box plot (or box-and-whisker plot) shows the distribution of quantitative data
in a way that facilitates comparisons between variables or across levels of a categorical variable.
The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution,
except for points that are determined to be outliers using a method that is a function of the inter-quartile range.
-->

````python
plt.figure(figsize=(15,5),facecolor='#00000000')
sns.boxplot(x = 'Age', y = 'Sex', data = datanew,hue ='Survived')
plt.legend( title="Survivor",loc=0, bbox_to_anchor=(0.55, 1.25),ncol=2, fancybox=True, shadow=True);
````

# Violin plot
Una trama de violn juega un papel similar a una trama de caja y bigotes.
Muestra la distribucin de datos cuantitativos en varios niveles de una (o ms) variables categricas.
de modo que esas distribuciones se puedan comparar.
A diferencia de un diagrama de caja, en el que todos los componentes del diagrama corresponden a puntos de datos reales,
la trama del violn presenta una estimacin de la densidad del ncleo de la distribucin subyacente.

Esta puede ser una forma efectiva y atractiva de mostrar mltiples distribuciones de datos a la vez,
pero tenga en cuenta que el procedimiento de estimacin est influenciado por el tamao de la muestra,
y los violines para muestras relativamente pequeas pueden parecer engaosamente suaves.
<!--
A violin plot plays a similar role as a box and whisker plot.
It shows the distribution of quantitative data across several levels of one (or more) categorical variables
such that those distributions can be compared. 
Unlike a box plot, in which all of the plot components correspond to actual datapoints,
the violin plot features a kernel density estimation of the underlying distribution.

This can be an effective and attractive way to show multiple distributions of data at once,
but keep in mind that the estimation procedure is influenced by the sample size,
and violins for relatively small samples might look misleadingly smooth.
-->

````python
plt.figure(figsize=(15,5),facecolor='#00000000')
sns.violinplot(x = 'Sex', y = 'Pclass', data = datanew, hue ='Survived')
plt.legend(title="Survivor",loc='upper center',bbox_to_anchor=(0.5, 1.30),ncol=2, fancybox=True, shadow=True);
````

# Strip plot
Dibuja una grfica de dispersin donde una variable sea categrica.

Un diagrama de tiras se puede dibujar por s solo, pero tambin es un buen complemento para un diagrama de caja o violn.
en los casos en los que desee mostrar todas las observaciones junto con alguna representacin de la distribucin subyacente.
<!--
Draw a scatterplot where one variable is categorical.

A strip plot can be drawn on its own, but it is also a good complement to a box or violin plot 
in cases where you want to show all observations along with some representation of the underlying distribution.
-->

````python
plt.figure(figsize=(15,5),facecolor='#00000000')
sns.stripplot(x = 'Pclass', y = 'Age', data = datanew,jitter=True,hue ='Survived',dodge=True )
plt.legend(["muerto","vivo"],loc=0);
````

# Swarm plot

Dibuja una grfica de dispersin categrica con puntos que no se superpongan.

Esta funcin es similar a stripplot (), pero los puntos se ajustan (solo a lo largo del eje categrico) para que no se superpongan. Esto proporciona una mejor representacin de la distribucin de valores, pero no se escala bien a un gran nmero de observaciones. Este estilo de trama a veces se denomina "enjambre de abejas".

Se puede dibujar un diagrama de enjambre por s solo, pero tambin es un buen complemento para un diagrama de caja o violn en los casos en que desee mostrar todas las observaciones junto con alguna representacin de la distribucin subyacente.

Organizar los puntos correctamente requiere una transformacin precisa entre los datos y las coordenadas de los puntos. Esto significa que se deben establecer lmites de eje no predeterminados antes de dibujar el grfico.
<!--
Draw a categorical scatterplot with non-overlapping points.

This function is similar to stripplot(), but the points are adjusted (only along the categorical axis) so that they dont overlap. This gives a better representation of the distribution of values, but it does not scale well to large numbers of observations. This style of plot is sometimes called a beeswarm.

A swarm plot can be drawn on its own, but it is also a good complement to a box or violin plot in cases where you want to show all observations along with some representation of the underlying distribution.

Arranging the points properly requires an accurate transformation between data and point coordinates. This means that non-default axis limits must be set before drawing the plot.
-->

````python
plt.figure(figsize=(15,5),facecolor='#00000000')
sns.violinplot(x = 'Pclass', y = 'Survived', data = datanew[0:100] ,hue ='Embarked')
sns.swarmplot(x = 'Pclass', y = 'Survived', data = datanew[0:100],hue ='Sex', color="red")
plt.legend(title="Embarkment port and Sex",loc='upper center', bbox_to_anchor=(0.5, 1.30),
          ncol=5, fancybox=True, shadow=True);
````

# Point plot

Muestra estimaciones puntuales e intervalos de confianza mediante glifos de grficos de dispersin.

Una grfica de puntos representa una estimacin de la tendencia central de una variable numrica por la posicin de los puntos de la grfica de dispersin y proporciona alguna indicacin de la incertidumbre en torno a esa estimacin utilizando barras de error.

Los grficos de puntos pueden ser ms tiles que los grficos de barras para enfocar las comparaciones entre diferentes niveles de una o ms variables categricas. Son particularmente hbiles para mostrar interacciones: cmo la relacin entre los niveles de una variable categrica cambia a travs de los niveles de una segunda variable categrica. Las lneas que unen cada punto del mismo nivel de tonalidad permiten juzgar las interacciones por las diferencias de pendiente, lo que es ms fcil para los ojos que comparar las alturas de varios grupos de puntos o barras.

Es importante tener en cuenta que una grfica de puntos muestra solo el valor medio (u otro estimador), pero en muchos casos puede ser ms informativo mostrar la distribucin de valores en cada nivel de las variables categricas. En ese caso, otros enfoques, como una trama de caja o violn, pueden ser ms apropiados.
<!--
Show point estimates and confidence intervals using scatter plot glyphs.

A point plot represents an estimate of central tendency for a numeric variable by the position of scatter plot points and provides some indication of the uncertainty around that estimate using error bars.

Point plots can be more useful than bar plots for focusing comparisons between different levels of one or more categorical variables. They are particularly adept at showing interactions: how the relationship between levels of one categorical variable changes across levels of a second categorical variable. The lines that join each point from the same hue level allow interactions to be judged by differences in slope, which is easier for the eyes than comparing the heights of several groups of points or bars.

It is important to keep in mind that a point plot shows only the mean (or other estimator) value, but in many cases it may be more informative to show the distribution of values at each level of the categorical variables. In that case, other approaches such as a box or violin plot may be more appropriate.
-->

````python
plt.figure(figsize=(15,5),facecolor='#00000000')
sns.pointplot(x = 'Pclass', y = 'Age', data = datanew,jitter=True,hue ='Survived',dodge=True )
plt.legend(title="Survived",loc=0, ncol=2, fancybox=True, shadow=True);
````

# Hist Plot (Histograma)

Traza histogramas univariados o bivariados para mostrar distribuciones de conjuntos de datos.

Un histograma es una herramienta de visualizacin clsica que representa la distribucin de una o ms variables contando el nmero de observaciones que se encuentran dentro de los contenedores de distribucin.

Esta funcin puede normalizar la estadstica calculada dentro de cada intervalo para estimar la frecuencia, densidad o masa de probabilidad, y puede agregar una curva suave obtenida usando una estimacin de densidad de kernel, similar a kdeplot ().
<!--
Plot univariate or bivariate histograms to show distributions of datasets.

A histogram is a classic visualization tool that represents the distribution of one or more variables by counting the number of observations that fall within disrete bins.

This function can normalize the statistic computed within each bin to estimate frequency, density or probability mass, and it can add a smooth curve obtained using a kernel density estimate, similar to kdeplot().
-->

````python
plt.figure(figsize=(15,5),facecolor='#00000000')
sns.histplot(x = 'Sex', y = 'Age', data = datanew,hue ='Pclass', palette=customPalette[0:3] );
````

# KDE plot  (Kernel Density Estimate)
Trace distribuciones univariadas o bivariadas utilizando la estimacin de densidad de kernel.

Una grfica de estimacin de densidad de kernel (KDE) es un mtodo para visualizar la distribucin de observaciones en un conjunto de datos, anlogo a un histograma. KDE representa los datos utilizando una curva de densidad de probabilidad continua en una o ms dimensiones.
El enfoque se explica con ms detalle en la
     [guia del usuario](https://seaborn.pydata.org/tutorial/distributions.html#tutorial-kde "Seaborn docs. ") <br>
En relacin con un histograma, KDE puede producir un grfico menos desordenado y ms interpretable, especialmente cuando se dibujan distribuciones mltiples. Pero tiene el potencial de introducir distorsiones si la distribucin subyacente es limitada o no uniforme. Como un histograma, la calidad de la representacin tambin depende de la seleccin de buenos parmetros de suavizado.
<!--
Plot univariate or bivariate distributions using kernel density estimation.

A kernel density estimate (KDE) plot is a method for visualizing the distribution of observations in a dataset, analagous to a histogram. KDE represents the data using a continuous probability density curve in one or more dimensions.
The approach is explained further in the 
    [user guide](https://seaborn.pydata.org/tutorial/distributions.html#tutorial-kde "Seaborn docs. ").<br>
Relative to a histogram, KDE can produce a plot that is less cluttered and more interpretable, especially when drawing multiple distributions. But it has the potential to introduce distortions if the underlying distribution is bounded or not smooth. Like a histogram, the quality of the representation also depends on the selection of good smoothing parameters.
-->

````python
plt.figure(figsize=(8,8),facecolor='#00000000')
sns.kdeplot(x = 'Age', y = 'Pclass', data = datanew,hue ='Survived', palette=['blue','red'] );
````

# ECDF plot<br> (empirical cumulative distribution functions)

Un ECDF representa la proporcin o recuento de observaciones que caen por debajo de cada valor nico en un conjunto de datos. En comparacin con un histograma o un grfico de densidad, tiene la ventaja de que cada observacin se visualiza directamente, lo que significa que no hay parmetros de agrupamiento o suavizado que deban ajustarse. Tambin ayuda a las comparaciones directas entre mltiples distribuciones. Una desventaja es que la relacin entre la apariencia de la trama y las propiedades bsicas de la distribucin (como su tendencia central, varianza y la presencia de cualquier bimodalidad) puede no ser tan intuitiva.

Ms informacin en la [gua del usuario](https://seaborn.pydata.org/tutorial/distributions.html#tutorial-ecdf "Seaborn docs."). <br>.
<!--
An ECDF represents the proportion or count of observations falling below each unique value in a dataset. Compared to a histogram or density plot, it has the advantage that each observation is visualized directly, meaning that there are no binning or smoothing parameters that need to be adjusted. It also aids direct comparisons between multiple distributions. A downside is that the relationship between the appearance of the plot and the basic properties of the distribution (such as its central tendency, variance, and the presence of any bimodality) may not be as intuitive.

More information is provided in the [user guide](https://seaborn.pydata.org/tutorial/distributions.html#tutorial-ecdf "Seaborn docs. ").<br>.
-->


````python
plt.figure(figsize=(8,8),facecolor='#00000000')
sns.ecdfplot(x = 'Age', data = datanew, hue='Pclass'  );
````

# Pair plot

Traza relaciones por pares en un conjunto de datos.

De forma predeterminada, esta funcin crear una cuadrcula de ejes de modo que cada variable numrica en los datos se compartir entre los ejes y en una sola fila y los ejes x en una sola columna. Las grficas diagonales se tratan de manera diferente: se dibuja una grfica de distribucin univariante para mostrar la distribucin marginal de los datos en cada columna.

Tambin es posible mostrar un subconjunto de variables o graficar diferentes variables en las filas y columnas.

<!--Plot pairwise relationships in a dataset.

By default, this function will create a grid of Axes such that each numeric variable in data will by shared across the y-axes across a single row and the x-axes across a single column. The diagonal plots are treated differently: a univariate distribution plot is drawn to show the marginal distribution of the data in each column.

It is also possible to show a subset of variables or plot different variables on the rows and columns.-->
